```python
import pandas as pd
import folium
cordinates=pd.read_csv("/content/data.txt",header=None)
cordinates.columns=['S.no','cam','city','state','plate']
cordinates.set_index('S.no',inplace=True)
cordinates['Address']=cordinates['city']+','+cordinates['state']
print(cordinates)

cordinates['coordinates']=cordinates['Address'].apply(nom.geocode)
cordinates['lat']=cordinates['coordinates'].apply(lambda x:x.latitude)
cordinates['lon']=cordinates['coordinates'].apply(lambda x:x.longitude)
cordinates
```
